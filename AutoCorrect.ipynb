### Import Libraries

```python
import pandas as pd
import numpy as np
import os
import re
import textdistance
from collections import Counter
```

### Read File & convert to list of words

```python
words = []
with open('auto.txt', 'r') as f:
    file_name_data = f.read()
#     print(file_name_data)
    file_name_data=file_name_data.lower()
#     print(file_name_data)
    words = re.findall('\w+',file_name_data)
    
words 
V = set(words)
```

```python
print(f"Top ten words in the text are:{words[0:10]}")
print(f"Total Unique words are {len(V)}.")
```

### cacluate Frequency of each word

```python
word_freq = {}
word_freq = Counter(words)
```

```python
word_freq
```

### find proabilties of each word

```python
probs = {}
total = sum(word_freq.values())
for k in word_freq.keys():
    probs[k]=word_freq[k]/total
```

```python
probs
```

### 

```python
def my_autocorrect(input_word):
    input_word = input_word.lower()
    
    if input_word in V:
            return('Your word seems to be correct')
    else:
        sim = [1-(textdistance.Jaccard(qval=2).distance(v,input_word)) for v in word_freq.keys()]
        df = pd.DataFrame.from_dict(probs, orient='index').reset_index()
        df = df.rename(columns={'index':'Word', 0:'Prob'})
        df['Similarity'] = sim
        output = df.sort_values(['Similarity', 'Prob'], ascending=False).head()
        return(output)
```
